Considere uma planilha onde foram anotadas as horas de estudos e as respectivas notas da prova. Verificar se os dados dos grupos seguem uma distribuição normal. Considerar um índice de significância de 5%.


Para verificar se um conjunto de dados segue uma distribuição normal em Python, você pode usar o teste de normalidade de Shapiro-Wilk. Aqui está um exemplo de programa em Python que realiza esse teste:

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2

# Carregue os dados do arquivo Excel com duas colunas
dados_excel = pd.read_excel('horas-notas.xlsx')

# Selecione as duas colunas de dados que você deseja analisar
coluna1 = dados_excel['Horas_estudos']
coluna2 = dados_excel['Notas']


# Realize o teste de normalidade de Shapiro-Wilk para cada coluna
stat1, p_valor1 = stats.shapiro(coluna1)
stat2, p_valor2 = stats.shapiro(coluna2)

# Defina um nível de significância (geralmente 0,05) para avaliar o p-valor
nivel_significancia = 0.05

# Crie um gráfico Q-Q comparando os dados com a distribuição normal
stats.probplot(coluna1, dist="norm", plot=plt)

# Adicione rótulos e um título ao gráfico
plt.xlabel("Quantis teóricos")
plt.ylabel("Quantis dos dados")
plt.title("Gráfico Q-Q - Coluna 1")

# Exiba o gráfico
plt.show()

# Crie um gráfico Q-Q comparando os dados com a distribuição normal
stats.probplot(coluna2, dist="norm", plot=plt2)

# Adicione rótulos e um título ao gráfico
plt2.xlabel("Quantis teóricos")
plt2.ylabel("Quantis dos dados")
plt2.title("Gráfico Q-Q - Coluna 2")

# Exiba o gráfico
plt2.show()


# Exiba os resultados do teste para a primeira coluna
print(f"Valor p da primeira coluna: {p_valor1}")
if p_valor1 > nivel_significancia:
    print("A primeira coluna segue uma distribuição normal.")
else:
    print("A primeira coluna não segue uma distribuição normal.")

# Exiba os resultados do teste para a segunda coluna
print(f"Valor p da segunda coluna: {p_valor2}")
if p_valor2 > nivel_significancia:
    print("A segunda coluna segue uma distribuição normal.")
else:
    print("A segunda coluna não segue uma distribuição normal.")

```
