```python
import matplotlib.pyplot as plt
import numpy as np

# Defina seus dados em uma lista
dados = [903.88, 1020.7 ,934.52 ,860.41 ,936.78 ,1036.92 ,915.38 ,1214.08, 1039.19 ,1086.98 ,1098.04 ,1014.53 ,993.45 ,950.38 ,1144.94 ,1011.26 ,1097.79 ,1120.19 ,941.83, 1066.12]


# Crie o gráfico de boxplot
plt.boxplot(dados)

# Calcule os principais pontos
minimo = np.min(dados)
maximo = np.max(dados)
mediana = np.median(dados)
media = np.mean(dados)
q1 = np.percentile(dados, 25)
q3 = np.percentile(dados, 75)

# Defina as cores para os pontos
cor_minimo = 'red'
cor_maximo = 'red'
cor_mediana = 'orange'
cor_quartis_1 = 'blue'
cor_quartis_3 = 'black'
cor_media = 'green'

# Adicione pontos no gráfico com as cores especificadas
plt.plot(1, minimo, 'o', color=cor_minimo, label='Mínimo')
plt.plot(1, maximo, 'o', color=cor_maximo, label='Máximo')
plt.plot(1, mediana, 'o', color=cor_mediana, label='Mediana')
plt.plot(1, media, 'o', color=cor_media, label='Média')
plt.plot(1, q1, 'o', color=cor_quartis_1, label='1º Quartil')
plt.plot(1, q3, 'o', color=cor_quartis_3, label='3º Quartil')


# Personalize rótulos
plt.xticks([1], ['Dados'])
plt.legend()

# Salve o gráfico em um arquivo de imagem (por exemplo, PNG)
plt.savefig("Exemplo01-boxplot.png")

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Defina seus dados em uma lista
grupo1 = [75.27,75.33,74.58,75.01,75.71,74.93,74.72,74.53,75.32,74.05]
grupo2 = [74.94,75.25,75.44,74.62,75.35,74.75,74.65,74.94,74.92,75.46]
grupo3 = [75.93,76.95,75.47,73.6,74.85,73.34,74.04,75,76.18,75.33]
grupo4 = [75.98,75.61,74.2,76.44,76.84,76.75,76.78,74.74,72.58,72.86]

# Criar uma lista de grupos para o boxplot
dados = [grupo1, grupo2, grupo3, grupo4]

# Criar um boxplot para comparar os grupos
plt.figure(figsize=(8, 6))
plt.boxplot(dados, labels=['Grupo 1', 'Grupo 2', 'Grupo 3', 'Grupo 4'])
plt.title('Boxplot para Comparar Grupos')
plt.xlabel('Grupos')
plt.ylabel('Valores')
plt.grid(True)

# Adicionar rótulos de quartis
plt.text(1, np.percentile(grupo1, 25), 'Q1', ha='center', va='center', color='red')
plt.text(1, np.percentile(grupo1, 50), 'Mediana', ha='center', va='center', color='red')
plt.text(1, np.percentile(grupo1, 75), 'Q3', ha='center', va='center', color='red')

plt.text(2, np.percentile(grupo2, 25), 'Q1', ha='center', va='center', color='red')
plt.text(2, np.percentile(grupo2, 50), 'Mediana', ha='center', va='center', color='red')
plt.text(2, np.percentile(grupo2, 75), 'Q3', ha='center', va='center', color='red')

plt.text(3, np.percentile(grupo3, 25), 'Q1', ha='center', va='center', color='red')
plt.text(3, np.percentile(grupo3, 50), 'Mediana', ha='center', va='center', color='red')
plt.text(3, np.percentile(grupo3, 75), 'Q3', ha='center', va='center', color='red')

plt.text(4, np.percentile(grupo4, 25), 'Q1', ha='center', va='center', color='red')
plt.text(4, np.percentile(grupo4, 50), 'Mediana', ha='center', va='center', color='red')
plt.text(4, np.percentile(grupo4, 75), 'Q3', ha='center', va='center', color='red')


# Salve o gráfico em um arquivo de imagem (por exemplo, PNG)
plt.savefig("boxplot-grupos.png")

plt.show()
```
