```python
import matplotlib.pyplot as plt

# Dados de exemplo
portos = ["Porto A", "Porto B", "Porto C", "Porto D"]
navios_ancorados = [25000, 30000, 20000, 15000]
navios_em_espera = [10000, 15000, 8000, 5000]
navios_partidos = [5000, 10000, 7000, 3000]
cargas_descarregadas = [30000, 45000, 35000, 25000]
cargas_embarcadas = [20000, 30000, 25000, 18000]

# Configuração do gráfico
width = 0.15  # Largura das barras
x = range(len(portos))

plt.bar(x, navios_ancorados, width, label='Navios Ancorados')
plt.bar([i + width for i in x], navios_em_espera, width, label='Navios em Espera')
plt.bar([i + 2 * width for i in x], navios_partidos, width, label='Navios Partidos')
plt.bar([i + 3 * width for i in x], cargas_descarregadas, width, label='Cargas Descarregadas')
plt.bar([i + 4 * width for i in x], cargas_embarcadas, width, label='Cargas Embarcadas')

plt.xlabel('Portos')
plt.ylabel('Quantidade')
plt.title('Estatísticas de Portos e Operações Navais')
plt.xticks([i + 2 * width for i in x], portos)
plt.legend()

# Exibição do gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
portos = ["Porto A", "Porto B", "Porto C", "Porto D"]
navios_ancorados = [25000, 30000, 20000, 15000]
navios_em_espera = [10000, 15000, 8000, 5000]
navios_partidos = [5000, 10000, 7000, 3000]
cargas_descarregadas = [30000, 45000, 35000, 25000]
cargas_embarcadas = [20000, 30000, 25000, 18000]

# Configuração do gráfico
width = 0.4
x = np.arange(len(portos))

fig, ax = plt.subplots()

ax.bar(x, navios_ancorados, width, label='Navios Ancorados')
ax.bar(x, navios_em_espera, width, label='Navios em Espera', bottom=navios_ancorados)
ax.bar(x, navios_partidos, width, label='Navios Partidos', bottom=np.array(navios_ancorados) + np.array(navios_em_espera))
ax.bar(x, cargas_descarregadas, width, label='Cargas Descarregadas', bottom=np.array(navios_ancorados) + np.array(navios_em_espera) + np.array(navios_partidos))
ax.bar(x, cargas_embarcadas, width, label='Cargas Embarcadas', bottom=np.array(navios_ancorados) + np.array(navios_em_espera) + np.array(navios_partidos) + np.array(cargas_descarregadas))

ax.set_xlabel('Portos')
ax.set_ylabel('Quantidade')
ax.set_title('Estatísticas de Portos e Operações Navais')
ax.set_xticks(x)
ax.set_xticklabels(portos)
ax.legend()

# Exibição do gráfico
plt.tight_layout()
plt.show()
```
