```python
import matplotlib.pyplot as plt
import numpy as np

# Gere dados aleatórios
dados = [2,2,3,10,13,14,15,15,16,16,18,18,20,21,22,22,23,24,25,25,26,27,29,29,30,32,36,42,44,45,45,46,48,52,58,59,61,61,61,65,66,66,68,75,78,80,89,90,92,97]

# Crie um histograma
plt.hist(dados, bins=15, edgecolor='black', alpha=0.8, color='red')

# Defina rótulos e título
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Exemplo de Histograma')

# Salve o gráfico em um arquivo de imagem (por exemplo, PNG)
plt.savefig("Exemplo02-histograma.png")

# Exiba o histograma
plt.show()
```

Neste exemplo:

Importamos a biblioteca Matplotlib como plt e a biblioteca NumPy como np.
Geramos dados aleatórios a partir de uma distribuição normal com média 0 e desvio padrão 1 usando np.random.normal().
Criamos o histograma com plt.hist(). Os argumentos bins especificam o número de barras no histograma, edgecolor define a cor das bordas das barras e alpha controla a transparência das barras.
Definimos rótulos para os eixos x e y usando plt.xlabel() e plt.ylabel().
Definimos um título para o histograma com plt.title().
Exibimos o histograma com plt.show().
