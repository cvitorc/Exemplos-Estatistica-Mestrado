Para verificar se um conjunto de dados segue uma distribuição normal em Python, você pode usar o teste de normalidade de Shapiro-Wilk. Aqui está um exemplo de programa em Python que realiza esse teste:

Neste código:

Importamos as bibliotecas NumPy e SciPy para trabalhar com os dados e realizar o teste de normalidade.
Definimos seus dados na lista dados.
Usamos stats.shapiro(dados) para realizar o teste de normalidade de Shapiro-Wilk, que retorna uma estatística de teste e um valor p.
Definimos um nível de significância, que é geralmente 0,05.
Exibimos os resultados do teste, incluindo a estatística de teste e o valor p.
Comparamos o valor p com o nível de significância para determinar se os dados seguem uma distribuição normal.
Se o valor p for maior que o nível de significância, você pode concluir que os dados se assemelham a uma distribuição normal. Caso contrário, os dados podem não seguir uma distribuição normal.








gráfico Quantis-Quantis, geralmente abreviado como gráfico Q-Q (ou Q-Q plot), é uma ferramenta gráfica usada para avaliar se um conjunto de dados segue uma distribuição teórica, como a distribuição normal. Ele compara os quantis (ou percentis) dos seus dados com os quantis esperados de uma distribuição teórica.

A ideia por trás de um gráfico Q-Q é que, se os dados seguem a distribuição teórica, os pontos no gráfico Q-Q devem formar uma linha aproximadamente reta, com um desvio mínimo da linha. A inclinação da linha e seu desvio em relação a uma reta perfeitamente diagonal são indicativos da conformidade dos dados com a distribuição teórica.

Aqui está como criar um gráfico Q-Q em Python usando a biblioteca Matplotlib e SciPy:

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Defina seus dados
dados = [903.88, 1020.7,934.52,860.41,936.78,1036.92,915.38,1214.08,1039.19,1086.98,1098.04,1014.53,993.45,950.38,1144.94,1011.26,1097.79,1120.19,941.83,1066.12]

# numero_de_dados = len(dados)
# print("Número de dados na lista: ", numero_de_dados)

# Realize o teste de normalidade de Shapiro-Wilk
stat, p_valor = stats.shapiro(dados)

# Defina um nível de significância (geralmente 0,05) para avaliar o p-valor
nivel_significancia = 0.05

# Crie um gráfico Q-Q comparando os dados com a distribuição normal
stats.probplot(dados, dist="norm", plot=plt)

# Adicione rótulos e um título ao gráfico
plt.xlabel("Quantis teóricos")
plt.ylabel("Quantis dos dados")
plt.title("Gráfico Q-Q")

# Exiba o gráfico
plt.show()


# Exibe o resultado do teste
print(f"Valor de estatística: {stat}")
print(f"Valor p: {p_valor}")

# Verifique se os dados seguem uma distribuição normal
if p_valor > nivel_significancia:
    print("Os dados seguem uma distribuição normal.")
else:
    print("Os dados não seguem uma distribuição normal.")
```

 O teste t é usado para determinar se a média de uma amostra difere significativamente de um valor de referência (geralmente uma média hipotética ou conhecida). Vamos supor que você tenha uma amostra de dados e queira realizar um teste t para determinar se a média da amostra é significativamente diferente de um valor médio de referência.

 H0 – A média das alturas não mudou em relação à média do mês anterior
H1 – A média das alturas mudou em relação à média do mês anterior


```python
import numpy as np
from scipy import stats

# Defina seus dados
amostra = [903.88, 1020.7,934.52,860.41,936.78,1036.92,915.38,1214.08,1039.19,1086.98,1098.04,1014.53,993.45,950.38,1144.94,1011.26,1097.79,1120.19,941.83,1066.12]


# Defina a média de referência (a média hipotética que você deseja testar)
media_referencia = 1055

# Calcula a média
soma_dos_dados = sum(amostra)
numero_de_dados = len(amostra)
media_dados = soma_dos_dados / numero_de_dados

print("Média dos dados:", media_dados)

# Realize o teste t
t_stat, p_valor = stats.ttest_1samp(amostra, media_referencia)

# Defina um nível de significância (geralmente 0,05) para avaliar o p-valor
nivel_significancia = 0.05

# Exiba o resultado do teste
print(f"Estatística t: {t_stat}")
print(f"Valor p: {p_valor}")

# Verifique se a média da amostra difere significativamente da média de referência
if p_valor < nivel_significancia:
    print("A média da amostra é significativamente diferente da média de referência.")
else:
    print("Considerando o nível de significância indicado, não há evidências para concluir que a média da amostra difere da média de referência.")
```
