```python
#importar as bibliotecas - pacotes
import math
import matplotlib.pyplot as plt
```

```python
#Definindo os dados
meses = ["jan", "fev", "mar", "abr", "mai", "jun", "jul", "ago", "set", "out", "nov", "dez"]
serie1 = [0.26, 0.61, 0.98, 1.26, 1.37, 1.69, 2.57, 3.39, 4.67, 6.34, 7.78, 8.66]
serie2 = [0.26, 0.36, 0.38, 0.18, 0.21, 0.32, 0.57, 0.79, 1.87, 1.28, 1.23, 0.66]
```

```python
# construcao do grafico

figura = plt.figure(figsize=(15,10))

#linestyle
# pontilhado = "":"
# tracejado = "--
# continua = "-"
# traço-ponto = "-."

#color
# azul = "b"
# vermelho = "r"
# preto = "k"
# verde = "g"
# magenta = "m"
# amarelo = "y"

#nome do grafico e dos eixos
titulo = "Gráfico: meses X Série"
titulo_eixox = "Meses"
titulo_eixoy = "Série"

# 1o grafico
figura.add_subplot(2,1,1)
plt.plot(meses,serie1, linestyle="-.", linewidth = 3, color = "b", label = "Série 1")
plt.scatter(meses, serie1, marker="v", s=120, color = "b")
plt.plot(meses,serie2, linestyle="-", linewidth = 3, color = "r", label = "Série 2")
plt.scatter(meses, serie2, marker="D", s=120, color = "r")
plt.title(titulo)
plt.xlabel(titulo_eixox)
plt.ylabel(titulo_eixoy)
plt.grid("true")
plt.legend()
# 2o grafico serie 1 - gráfico de barra
figura.add_subplot(2,2,3)
plt.bar(meses,serie1, linewidth = 3, color = "b", label = "Série 1")
plt.xlabel(titulo_eixox)
plt.ylabel(titulo_eixoy)
plt.legend()

# 3o grafico serie 2 - gráfico de barra
figura.add_subplot(2,2,4)
plt.bar(meses,serie2, linewidth = 3, color = "r", label = "Série 2")
plt.xlabel(titulo_eixox)
plt.ylabel(titulo_eixoy)
plt.legend()

plt.show()
```
